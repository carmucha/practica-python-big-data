```python
from functools import reduce, partial

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']


# A. Una función que calcule el precio total de la compra
def precio_total(compra, articulos):
    precio_compra = reduce(lambda acc, item: articulos[item] + acc,  compra, 0)
    return precio_compra

print('Precio Compra: ', precio_total(compra, articulos))


# B. Una función que calcule el precio total de la compra con un descuento del 10%

def precio_con_descuento(porcetaje_descuento, precio):
    descuento = precio * porcetaje_descuento/100
    return precio - descuento

descuento_10_porciento = partial(precio_con_descuento, 10)

print("Descuento del 10%: ", descuento_10_porciento(precio_total(compra, articulos)))


# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€

def descuento_compra_mayor_100(total):
    if total > 100:
        result = descuento_10_porciento(total)
    else:
        result = total
    return result

print('Descuento compra mayor de 100: ', descuento_compra_mayor_100(precio_total(compra, articulos)))


# D. Una función que calcule el precio total aplicando el IVA (21%)

add_impuesto = lambda impuesto, cantidad: (cantidad * impuesto/100) + cantidad

add_iva_21 = partial(add_impuesto, 21)

calcula_total_mas_iva = add_iva_21(precio_total(compra, articulos))

print('Precio con IVA: ', calcula_total_mas_iva)

# E.lista los artículos cuyo precio es superior a 20€

menor_20 = list(filter(lambda item: articulos[item] > 20, compra))
print('Productos con precio menor de 20: ', menor_20)
```
